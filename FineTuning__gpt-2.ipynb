```python
# # Convert the DataFrame to a Dataset object
dataset = Dataset.from_pandas(combined_df[['text']])

# # Load the tokenizer and model
# tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
# model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
# Load the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
# Add a padding token to the tokenizer
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=50)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
# Training arguments
training_args = TrainingArguments(
    output_dir="./gpt2-bakery-model",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    save_steps=10_000,
    save_total_limit=2,
)
from transformers import DataCollatorForLanguageModeling
# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
)
```

```python
# Fine-tune the model
trainer.train()
```

```python
import torch

# Example of an incorrect product name
incorrect_name = "coca"

# Prepare the input text in the format used during training
input_text = f"incorrect_name: {incorrect_name} -> correct_name:"

# Tokenize the input
inputs = tokenizer.encode(input_text, return_tensors="pt")

# Move the input tensors to the GPU if available
if torch.cuda.is_available():
    inputs = inputs.to("cuda")

# Generate the output
outputs = model.generate(inputs, max_length=50,num_return_sequences=1)


# Decode the output to get the corrected name
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```
